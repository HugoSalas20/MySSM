```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
m1=35
k1=900
b1=30
```

```python
t=0
q0i=0
q1i=0
tsol=[t]
q1=[q0i]
q1p=[q1i]
t=0
```

```python
#generar el sistema
def FCircuito(t,y) :
    y0P=y[1]
    y1P=(k1/m1)*y[0]-(b1/m1)*y[1]
    return np.array([y0P,y1P])

#cambiar appends
def RK4 (Func,dt,Ti,Tf,y) :
    n=(Tf-Ti)/dt
    t=Ti
    t2=Ti
    x=n+t
    while t<x:
      k1=Func(t2,y)
      k2=Func(t2+dt/2,y+dt*k1/2)
      k3=Func(t2+dt/2,y+dt*k2/2)
      k4=Func(t2+dt,y+dt*k3)
      y=y+(k1+2*k2+2*k3+k4)*dt/6
      q1.append(y[0])
      q1p.append(y[1])

      t=t+1
      t2=t2+dt
      tsol.append(t2)
```

```python
y0=[q0i,q1i]
RK4 (FCircuito,0.01,0,10,y0)
plt.plot(tsol,q1)
plt.grid()
plt.show()

plt.plot(tsol,q1p)
plt.grid()
plt.show()
```
